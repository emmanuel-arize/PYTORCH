```python
import torch
import numpy as np
import torch.nn as nn
import os
import pandas as pd
from torch.autograd import Variable
```

```python
x_data=Variable(torch.Tensor([[2],[4],[6],[8],[10],[12]]))
y_data=Variable(torch.Tensor([[1],[3],[5],[7],[9],[11]]))
print("x and y are 5 by 1 matrices")
print(x_data)
print(y_data)
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        #self.linear=nn.Linear(input_dim,output_dim)
        self.linear=nn.Linear(1,1)# 1 in and 1 out
        
    def forward(self,x):
        y_pred=self.linear(x)
        return y_pred
        
        
model=Model()
'''
class torch.nn.MSELoss(size_average=True)
Creates a criterion that measures the mean squared error between n elements in the input x and target y:
'''
loss_function=nn.MSELoss(size_average=False)
#  stochastic gradient descent with a leaning rate (lr) of 0.05
optimizer=torch.optim.SGD(model.parameters(),lr=0.001,momentum=0.9)
for eporch in range(50):
    # predicting y
    y_pred=model(x_data)
    # computing the loss
    loss=loss_function(y_pred,y_data)
    print('epoch \n',eporch,'  y_values \n',y_data,'y_pred\n',y_pred,' ','loss \n',loss.data)
    # Zero gradients, perform a backward pass, and update the weights.
    
    # zero the gradients
    optimizer.zero_grad()
    #perform backward-propagation
    loss.backward()
    # All optimizers implement a step() method, that updates the parameters. It can be used as:
    #optimizer.step()
    optimizer.step()
    
#[w,b]=model.parameters
#def get_para():
 #   return w.data[0][0], b.data[0]

hour_four=Variable(torch.Tensor([[4.0]]))   

print('after training',4,model.forward(hour_four).data[0][0])
```

```python
x_train_data=data.loc[:,['Temperature','Rainfall','Flyers','Price']]
y_train_data=data.loc[:,'Sales']
```

```python
#y=torch.Tensor(y_train_data)
#y_train_data=np.array(y_train_data)
x=np.array(x_train_data)
y=np.array(y_train_data)
y=torch.from_numpy(y)
x=torch.from_numpy(x)
```

```python
x=Variable(x)
y=y.reshape(-1,1)
y=Variable(y)
```

```python
class L_regression(nn.Module):
    def __init__(self,n_features):
        super(L_regression,self).__init__()
        #self.linear=nn.Linear(input_dim,output_dim)
        self.linear=nn.Linear(in_features=n_features,out_features=1,bias=True)
        
    def forward(self,x_in):
        return self.linear(x_in)
        
l_reg=L_regression(n_features=4)
'''
class torch.nn.MSELoss(size_average=True)
Creates a criterion that measures the mean squared error between n elements in the input x and target y:
'''
loss_fu=nn.MSELoss()
#  stochastic gradient descent with a leaning rate (lr) of 0.05
#opt=torch.optim.SGD(l_reg.parameters(),lr=0.01,momentum=0.7)
opt=torch.optim.Adam(l_reg.parameters(),lr=0.01,weight_decay=1e-5)


for eporch in range(100):
    # predicting y
    y_pred=l_reg(x.float())
    # computing the loss
    loss=loss_fu(y_pred,y.float())
    print('epoch \n',eporch,'\n loss \n',loss.data)
    # Zero gradients, perform a backward pass, and update the weights.
    
    # zero the gradients
    opt.zero_grad()
    #perform backward-propagation
    loss.backward()
    # All optimizers implement a step() method, that updates the parameters. It can be used as:
    #optimizer.step()
    opt.step()
```
